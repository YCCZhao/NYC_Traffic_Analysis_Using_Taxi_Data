# Helper Functions

```python
#https://github.com/am2786/NYC-taxi-data-analysis/blob/master/Sections%204%20%26%205.ipynb
us_holidays = holidays.UnitedStates()

#geo_data =  json.load(open('NY_neighborhoods.geojson'))   # GeoJson file
geo_data =  json.load(open('NY_manhanttan_neighborhoods.geojson'))  
area_data = json.load(open('area_dict.json'))  

def holiday(datetime):
    return datetime in us_holidays

def day_of_week(datetime):
    return datetime.isoweekday()

def neighborhood_mapper(lon, lat):
    try:
        for i, feature in enumerate(geo_data['features']):     
            coords = feature['geometry']['coordinates'][0]
            p = path.Path(coords)
            if p.contains_point((lon, lat)):
                return feature['properties']['neighborhood']
    except:
        return 'none'
    return 'other'
        
def vendor_id(code):
    if code == 1:
        return "CMT"
    elif code == 2:
        return "VFI"
    else:
        return "other"

def area_mapper(neighborhood):
    try:
        return area_data[neighborhood]
    except:
        return neighborhood
# udf converts a function into one that can be applied over
# a dataframe column. This is kind of like Pandas apply/map functionality.
holiday = udf(holiday)
day_of_week = udf(day_of_week)
neighborhood_mapper = udf(neighborhood_mapper)
vendor_id = udf(vendor_id)
area_mapper = udf(area_mapper)
```

```python
def taxi_data_cleaner(data, to_datetime, to_vendor_code, payment_array):
    
    # Getting rid of rides with 0 passengers
    data = data.filter(data.passenger_count > 0)
    data = data.filter(data.passenger_count <= 6.)
    
    # Getting rid of rides with impossibly low fares
    data = data.filter(data.fare_amount > 1.)
    
    # Getting rid of rides with impossibly low duration
    data = data.filter(lower(data.payment_type).isin(payment_array) == True)  
    
    # Getting rid of rides with impossibly low duration
    data = data.filter(data.trip_distance <= 60.)
    data = data.filter(data.trip_distance > 0.)

    if to_datetime:
        data = data.withColumn("pickup_datetime", 
                               to_timestamp('pickup_datetime', 'MM/dd/yyyy hh:mm:ss a'))
        data = data.withColumn("dropoff_datetime",
                               to_timestamp('dropoff_datetime', 'MM/dd/yyyy hh:mm:ss a'))
    
    #if to_vendor_code:
    #    data = data.withColumn("vendor_id", vendor_id('vendor_id'))
    
    data = data.withColumn("trip_duration", (unix_timestamp('dropoff_datetime', 'MM/dd/yyyy hh:mm:ss a') 
                                        - unix_timestamp('pickup_datetime', 'MM/dd/yyyy hh:mm:ss a'))
                           .cast('double'))
    
    data = data.withColumn("pickup_year", year('pickup_datetime'))
    #data = data.withColumn("dropoff_year", year('dropoff_datetime'))
    data = data.withColumn("pickup_month", month('pickup_datetime'))
    #data = data.withColumn("dropoff_month", month('dropoff_datetime'))
    data = data.withColumn("pickup_day", dayofmonth('pickup_datetime'))
    data = data.withColumn("pickup_dayweek", day_of_week('pickup_datetime')) 
    data = data.withColumn("pickup_hour", hour('pickup_datetime'))
    #data = data.withColumn("dropoff_hour", hour('dropoff_datetime'))
    data = data.withColumn("is_holiday", holiday('pickup_datetime'))   
    data = data.withColumn("PU_neighborhood", neighborhood_mapper('pickup_longitude', 'pickup_latitude')) 
    data = data.withColumn("PU_area", area_mapper('PU_neighborhood')) 
    data = data.withColumn("DO_neighborhood", neighborhood_mapper('dropoff_longitude', 'dropoff_latitude'))
    data = data.withColumn("DO_area", area_mapper('DO_neighborhood')) 
    data = data.withColumn("trip_speed", 
                           (data.trip_distance.cast('double')) / (data.trip_duration.cast('double')/3600))
    
    # Getting rid of rides with impossibly low duration
    data = data.filter(data.trip_duration > 0.)
    data = data.filter(data.trip_duration <= 3600.)
    
    # Getting rid of rides with impossibly low duration
    data = data.filter(data.trip_speed <= 60.)
    
    return(data)
```

```python
new_col_name = {'yellow': [new_col_yl_bf16, new_col_yl_bf16],
                'green': [new_col_gr_13, new_col_gr_14, 
                          new_col_gr_15, new_col_gr_16]}

yellow_data = [('yellow', x, False, False) for x in range(2009, 2016)] + [('yellow', 2016, True, True)]
green_data = [('green', x, True, True) for x in range(2013, 2017)]
#both_data = green_data + [('yellow', 2016, True, True)] + yellow_data
```

```python
%%time
for data_point in [yellow_data[4]]:
    taxi_type = data_point[0]
    data_year = data_point[1]
    to_datetime = data_point[2]
    to_vendor_id = data_point[3]
    
    # file path
    file = "s3://ycczhao-nyc-taxi-data/" + str(data_year) + '_' + taxi_type + '.csv'
    # read in as dataframe
    df = sqlContext.read.csv(file, sep=',',
                         inferSchema=True,
                         header=True)
    
    # change column names
    if taxi_type == 'yellow' and data_year < 2016:
        to_col_name = new_col_name[taxi_type][0]
    elif taxi_type == 'yellow' and data_year == 2016:
        to_col_name = new_col_name[taxi_type][1]
    else:
        to_col_name = new_col_name[taxi_type][data_year-2013]
        
    for old, new in zip(df.columns, to_col_name):
        df = df.withColumnRenamed(old, new)

        
    # drop useless columns
    for field in ['store_and_forward', 
                  'trip_type', 'vendor_id',
                  'mta_tax', 'tip_amount',
                  'total_amount','PULocationID', 'DOLocationID',
                  'Ehail_fee', 'improvement_surcharge', 
                  'extra', 'tolls_amount',
                  'Improvement_surcharge']:

        df = df.drop(field)
            
    # payment codes vary every year    
    if taxi_type == 'yellow':
        payment_code = payment_dict[data_year]
    else:
        payment_code = payment_dict[taxi_type]

    # data cleaning    
    new_df = taxi_data_cleaner(df, to_datetime, to_vendor_id, payment_code)

    # drop useless columns
    for field in ['pickup_longitude', 'pickup_latitude',
                  'dropoff_latitude', 'dropoff_longitude',
                  'fare_amount', 'payment_type',
                  'passenger_count']:

        new_df = new_df.drop(field)
        
```

```python
output_file = "s3://ycczhao-nyc-taxi-data/new_" + str(data_year) + '_' + taxi_type + '.parquet'
new_df.repartition(48).write.partitionBy('PU_area', 'DO_area').parquet(output_file, mode="overwrite")
```
